```python
import torch
import pandas as pd
import numpy as np

from fc_layer import FCLayer 
from act_layer import ActivationLayer
from activation import Activation
from activation_prime import ActivationPrime
from loss import Loss
from loss_prime import LossPrime
from network import Network
```

Đọc dữ liệu từ file data.csv

```python
x_train = pd.read_csv('data.csv', usecols=[0,1,2]) # Read first 3 variable columns in dataset
x_train = x_train.values.tolist() # Convert dataframe to list

y_train = pd.read_csv('data.csv', usecols=[3]) # Read result columns in dataset
y_train = y_train.values.tolist() # Convert dataframe to list
```

Dữ liệu của file data.csv

```python
for i in range(len(x_train)):
    print("#"+str(i+1)+": "+str(x_train[i])+" "+str(y_train[i]))
```

Set up dữ liệu để train

```python
# training data
x_trainset = list()
y_trainset = list()
for i in range(80):
    x_temp = list()
    x_temp.append(x_train[i])
    x_max = max(x_train[i])
    x_edit = [[x / x_max for x in x_temp[0]]]
    x_trainset.append(x_edit)

    y_temp = list()
    y_temp.append(y_train[i])
    y_edit = [[y / 100 for y in y_temp[0]]]
    y_trainset.append(y_edit)

x_set = torch.tensor(x_trainset, dtype=torch.float)
y_set = torch.tensor(y_trainset, dtype=torch.float)
```

Set up mạng neural<br>
Bao gồm: <br>
+) Input layer: 3 perceptrons <br>
+) 1st Hidden layers: 15 perceptrons <br>
+) 2nd Hidden layers: 12 perceptrons <br>
+) Output layer: 1 perceptron <br>

```python
# network architecture with tanh activation
net = Network()
net.add(FCLayer(3, 15))
net.add(ActivationLayer(Activation.tanh, ActivationPrime.tanh_derivative))
net.add(FCLayer(15, 12))
net.add(ActivationLayer(Activation.tanh, ActivationPrime.tanh_derivative))
net.add(FCLayer(12, 1))
net.add(ActivationLayer(Activation.tanh, ActivationPrime.tanh_derivative))

# train your network
net.use(Loss.mse, LossPrime.mse_prime)
net.fit(x_set, y_set, epochs=1400, alpha=0.01) # learning rate = 0.01
```

Prepare the dataset to test (included in data.csv)

```python
def scaling(testcase):
    num = max(testcase)
    result = [[i / num for i in testcase]]
    return result
```

```python
# test
testing = list()
case_1 = [[94.2, 4.9, 8.1]] # correct result 14.0
case_2 = [[17.2,4.1,31.6]] # correct result 5.9
case_3 = [[56.2,5.7,29.7]] # correct result 8.7

testcase_1 = scaling(case_1[0])
testing.append(testcase_1)

testcase_2 = scaling(case_2[0])
testing.append(testcase_2)

testcase_3 = scaling(case_3[0])
testing.append(testcase_3)

for i in range(len(testing)):
    print(testing[i])
```

```python
# testing phase
data = torch.tensor(testing, dtype = torch.float)

out = net.predicts(data)
print(out)
```
